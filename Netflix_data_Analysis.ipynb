```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
print(f"Headers :{df.columns}")
df.head()
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['country'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['rating'].fillna('Not Rated', inplace=True)
```

```python
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
sns.countplot(x='type', data=df, palette=['orange', 'green'])
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['release_year'].value_counts().head(10)
```

```python
df['release_year'].value_counts().head(10).plot(kind='bar')
plt.title("Top Years with Most Releases")
plt.show()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
df['listed_in'].value_counts().head(10).plot(kind='barh')
plt.title("Top Genres on Netflix")
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(
    kind='line',
    marker='o',
    color='orange'
)

plt.title("Content Added to Netflix Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
sns.countplot(y='rating', data=df)
plt.title("Content Ratings Distribution")
plt.show()
```
