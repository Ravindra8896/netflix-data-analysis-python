```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
print(f"Headers :{df.columns}")
df.head()
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['country'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['rating'].fillna('Not Rated', inplace=True)
```

```python
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
sns.countplot(x='type', data=df, palette=['orange', 'green'])
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['release_year'].value_counts().head(10)
```

```python
df['release_year'].value_counts().head(10).plot(kind='bar')
plt.title("Top Years with Most Releases")
plt.show()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
df['listed_in'].value_counts().head(10).plot(kind='barh')
plt.title("Top Genres on Netflix")
plt.show()
```

```python
df.columns
```

```python
top_countries = df['country'].value_counts().head(10)

# Create Figure
plt.figure(figsize=(10,6))

# Plot Bar Chart
top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Netflix Content", fontsize=14, fontweight='bold')
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

# Save Image (VERY IMPORTANT LINE)
plt.savefig("images/top_countries.png", dpi=300, bbox_inches='tight')

# Show Plot
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(
    kind='line',
    marker='o',
    color='orange'
)

plt.title("Content Added to Netflix Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.savefig("Images/content_growth.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.countplot(y='rating', data=df)
plt.title("Content Ratings Distribution")
plt.savefig("Images/movies_vs_tvshows.png")
plt.show()
```
